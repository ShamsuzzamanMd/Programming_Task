```python
# Solution of No. 1

def examine(num):
    if num in range(11):
        if num>7:
            print("passed")
        elif num<5:
            num=num+1
            print(num)
            check(num)
        else:
            num=num+1
            print(num)
            return check(num)

num=int(input("a number between 0-10"))
examine(num)



# Solution of No. 2

def situation(num):
    return 1 if n<2 
    else num*situation(num-1)

num=int(input("a number"))
print(situation(num))


# Solution of No. 4

def Findtriplet(ar):
    if len(ar)<3:
        return False
    ar.sort()
    for i in range(len(ar)-2):
        if ar[i]+ar[i+1]>ar[i+2]:
            return True
def check(ar):
    if Findtriplet(ar):
        print(1)
    else:
        print(0)
        
ar = [10,2,5,1,8,20]
ar1 = [10,50,5,1]
check(ar)
check(ar1)



# Solution of No. 5

def MissedElement(ar):
    num=len(ar)
    total =(num+1)*(num+2)//2
    return total-sum(ar)

ar=[2,3,1,5]
print(MissedElement(ar))



# Solution of No. 6

def rotate_ar(a,k):
    return a[-k:]+a[:-k]
a=[3,8,9,7,6]
b=[0,0,0]
c=[1,2,3,4]
print( "{0} -> {1} ".format(a,rotate_ar(a,3)))
print( "{0} -> {1} ".format(b,rotate_ar(b,1)))
print( "{0} -> {1} ".format(c,rotate_ar(c,4)))
```
